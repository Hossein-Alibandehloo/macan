```python
import imp
from pptx import Presentation
from pptx.util import Inches
import os
from pptx.util import Pt
from PIL import Image
import pandas as pd
from os import listdir
from os.path import isfile, join
from pptx.dml.color import RGBColor
from pptx.enum.shapes import MSO_SHAPE 
import datetime, re, calendar
from datetime import datetime
from pptx.enum.shapes import MSO_CONNECTOR
from pptx.enum.text import PP_ALIGN

class Pptx_maker_telegram:
    logo_pic = 'macan_logo2.png'
    laptop_temp = 'laptop_temp.png'

    def add_slide(self, data, pptp):
        Layout = pptp.slide_layouts[1]
        pptp.slide_height = Inches(13)
        pptp.slide_width = Inches(24)
        slide = pptp.slides.add_slide(Layout)
        
        # slide.shapes.add_picture(self.logo_pic, Inches(0.2), Inches(12.5), width=Inches(1.72))
        if 'Insight' == data['type']:
            slide.shapes.add_picture(data['path'], 
            # Inches(4.2),
            Inches(10.2),
            Inches(2.1), 
            width=Inches(4.2),
            height=Inches(8)
            )
        else:
            slide.shapes.add_picture(data['path'], 
            Inches(7.1),
            Inches(0.9), 
            # width=Inches(4.2),
            height=Inches(10)
            )

        self.add_date(calendar.month_name[data['date'].month] + str(', ') + str(data['date'].day) + str(', ') + str(data['date'].year), slide)
        self.add_id(data['id'].replace("/", ''), slide)
        line = slide.shapes.add_connector(MSO_CONNECTOR.STRAIGHT, Inches(23), Inches(.75), Inches(23), Inches(1.75))
        
        line.line.color.rgb = RGBColor(127, 127, 127)
        self.add_platform("تلگرام ", slide)
        if data['type'] == 'Evidence':
            self.add_evidence_insight(' اویدنس', slide)
            line2 = slide.shapes.add_connector(MSO_CONNECTOR.STRAIGHT, Inches(10.8), Inches(11.13), Inches(10.8), Inches(12.13))
            line2.line.color.rgb = RGBColor(127, 127, 127)
        else:
            self.add_evidence_insight(" اینسایت استوری", slide)
            line2 = slide.shapes.add_connector(MSO_CONNECTOR.STRAIGHT, Inches(10.8), Inches(11.13), Inches(10.8), Inches(11.7))
            line2.line.color.rgb = RGBColor(127, 127, 127)
        
    def convert_date(self, date):
        inds = [i.start() for i in re.finditer('/', date)]
        d , m , y = '','',''
        for i in range(0, inds[0]):
            d += date[i]
        for i in range(inds[0] + 1, inds[1]):
            m += date[i]
        for i in range(inds[1] + 1, len(date)):
            y += date[i]
        jalili_date =  datetime.date.fromgregorian(day=int(d),month=int(m),year=int(y))
        
        return str(jalili_date).replace('-','.')



    def image_inserter2(self, address, slide):
        slide.shapes.add_picture(address,Inches(7.6),Inches(3), Inches(16),
        #  Inches(11.5)
        )
        
    def add_id(self, id, slide):
        t = slide.shapes.add_textbox(height=1, width=1, left=11, top=1)

        # t.text_frame.margin_left = Inches(0)
        # t.text_frame.margin_left = Inches(12 - 0.5 * len(id) * 0.14)
        t.text_frame.margin_left = Inches(11)
        t.width
        t.text_frame.margin_top = Inches(11.25)
        for paragraph in t.text_frame.paragraphs:
            paragraph.alignment = PP_ALIGN.LEFT
        p = t.text_frame.paragraphs[0]
        # p2 = t.text_frame.paragraphs[1]
        run = p.add_run()
        run.text = id
        font = run.font
        font.name = 'Yekan Bakh'
        font.size = Pt((18*40)/22.2)
        font.color.rgb = RGBColor(73, 31, 103)

    def add_date(self, date, slide):
        t = slide.shapes.add_textbox(height=40, width=5, left=10, top=5)
        # t.text_frame.margin_left = Inches(12 - 0.5 * len(date) * 0.19)
        t.text_frame.margin_left = Inches(11)
        t.text_frame.margin_top = Inches(11.75)

        p = t.text_frame.paragraphs[0]

        run = p.add_run()
        run.text = date
        font = run.font
        font.name = 'Yekan Bakh'
        font.size = Pt((16*40)/22.2)
        
        font.color.rgb = RGBColor(73, 31, 103)
    def add_evidence_insight(self, date, slide):
        t = slide.shapes.add_textbox(height=4, width=5, left=10, top=5)
        t.text_frame.margin_left = Inches(21.8)
        t.text_frame.margin_top = Inches(1.45)

        p = t.text_frame.paragraphs[0]
        run = p.add_run()
        run.text = date
        font = run.font
        font.name = 'Yekan Bakh'
        font.size = Pt((14*40)/22.2)
        font.color.rgb = RGBColor(166, 166, 166)
    def add_platform(self, date, slide):
        t = slide.shapes.add_textbox(height=4, width=5, left=10, top=5)
        t.text_frame.margin_left = Inches(21.8)
        t.text_frame.margin_top = Inches(.75)
        
        p = t.text_frame.paragraphs[0]
        run = p.add_run()
        run.text = date
        font = run.font
        font.name = 'Yekan Bakh'
        font.bold = True
        font.size = Pt((18*40)/22.2)
        font.color.rgb = RGBColor(73, 31, 103)

    def add_shape1(self, slide):
        shapes = slide.shapes
        left = top = width = height = Inches(1.05)
        shape = shapes.add_shape(
            MSO_SHAPE.RECTANGLE, left, top, width, height
        )
        shape.line.fill.background()

        fill = shape.fill
        fill.solid()
        fill.fore_color.rgb = RGBColor(237, 157, 45)
        shape.left = Inches(0)
        shape.top = Inches(.67)
    def add_shape2(self, slide):
        shapes = slide.shapes
        left = top = Inches(1.0)
        width = Inches(4.75)
        height = Inches(0.6)
        shape = shapes.add_shape(
            MSO_SHAPE.RECTANGLE, left, top, width, height
        )

        fill = shape.fill
        fill.solid()
        fill.fore_color.rgb = RGBColor(73, 31, 103)
        shape.left = Inches(19.25)
        shape.top = Inches(1.1)
        shape.line.fill.background()


class Pptx_maker_insta:
    logo_pic = 'macan_logo3.png'
    laptop_temp = 'laptop_temp.png'

    def add_slide(self, data, pptp):
        Layout = pptp.slide_layouts[1]
        pptp.slide_height = Inches(13)
        pptp.slide_width = Inches(24)
        slide = pptp.slides.add_slide(Layout)
        
        slide.shapes.add_picture(self.logo_pic, Inches(0.2), Inches(12.5), width=Inches(1.72))
        if 'Insight' == data['type']:
            slide.shapes.add_picture(data['path'], 
            # Inches(4.2),
            Inches(10.2),
            Inches(2.1), 
            width=Inches(4.2),
            height=Inches(8)
            )
        else:
            slide.shapes.add_picture(data['path'], 
            Inches(4.2),
            # Inches(7.2),
            Inches(2.1), 
            width=Inches(15.6),
            )

        self.add_date(calendar.month_name[data['date'].month] + str(', ') + str(data['date'].day) + str(', ') + str(data['date'].year), slide)

        self.add_id(data['id'].replace("/", ''), slide)
        line = slide.shapes.add_connector(MSO_CONNECTOR.STRAIGHT, Inches(23), Inches(.75), Inches(23), Inches(1.75))
        
        line.line.color.rgb = RGBColor(127, 127, 127)
        self.add_platform("اینستاگرام", slide)
        if data['type'] == 'Evidence':
            self.add_evidence_insight(' اویدنس', slide)
            line2 = slide.shapes.add_connector(MSO_CONNECTOR.STRAIGHT, Inches(10.8), Inches(11.13), Inches(10.8), Inches(12.13))
            line2.line.color.rgb = RGBColor(127, 127, 127)
        else:
            self.add_evidence_insight(" اینسایت استوری", slide)
            line2 = slide.shapes.add_connector(MSO_CONNECTOR.STRAIGHT, Inches(10.8), Inches(11.13), Inches(10.8), Inches(11.7))
            line2.line.color.rgb = RGBColor(127, 127, 127)
        
    def convert_date(self, date):
        inds = [i.start() for i in re.finditer('/', date)]
        d , m , y = '','',''
        for i in range(0, inds[0]):
            d += date[i]
        for i in range(inds[0] + 1, inds[1]):
            m += date[i]
        for i in range(inds[1] + 1, len(date)):
            y += date[i]
        jalili_date =  datetime.date.fromgregorian(day=int(d),month=int(m),year=int(y))
        
        return str(jalili_date).replace('-','.')



    def image_inserter2(self, address, slide):
        slide.shapes.add_picture(address,Inches(7.6),Inches(3), Inches(16),
        #  Inches(11.5)
        )
        
    def add_id(self, id, slide):
        t = slide.shapes.add_textbox(height=1, width=1, left=11, top=1)

        # t.text_frame.margin_left = Inches(0)
        # t.text_frame.margin_left = Inches(12 - 0.5 * len(id) * 0.14)
        t.text_frame.margin_left = Inches(11)
        t.width
        t.text_frame.margin_top = Inches(11.25)
        for paragraph in t.text_frame.paragraphs:
            paragraph.alignment = PP_ALIGN.LEFT
        p = t.text_frame.paragraphs[0]
        # p2 = t.text_frame.paragraphs[1]
        run = p.add_run()
        run.text = id
        font = run.font
        font.name = 'Yekan Bakh'
        font.size = Pt((18*40)/22.2)
        font.color.rgb = RGBColor(73, 31, 103)

    def add_date(self, date, slide):
        t = slide.shapes.add_textbox(height=40, width=5, left=10, top=5)
        # t.text_frame.margin_left = Inches(12 - 0.5 * len(date) * 0.19)
        t.text_frame.margin_left = Inches(11)
        t.text_frame.margin_top = Inches(11.75)

        p = t.text_frame.paragraphs[0]

        run = p.add_run()
        run.text = date
        font = run.font
        font.name = 'Yekan Bakh'
        font.size = Pt((16*40)/22.2)
        
        font.color.rgb = RGBColor(73, 31, 103)
    def add_evidence_insight(self, date, slide):
        t = slide.shapes.add_textbox(height=4, width=5, left=10, top=5)
        t.text_frame.margin_left = Inches(21.8)
        t.text_frame.margin_top = Inches(1.45)

        p = t.text_frame.paragraphs[0]
        run = p.add_run()
        run.text = date
        font = run.font
        font.name = 'Yekan Bakh'
        font.size = Pt((14*40)/22.2)
        font.color.rgb = RGBColor(166, 166, 166)
    def add_platform(self, date, slide):
        t = slide.shapes.add_textbox(height=4, width=5, left=10, top=5)
        t.text_frame.margin_left = Inches(21.22)
        t.text_frame.margin_top = Inches(.75)
        
        p = t.text_frame.paragraphs[0]
        run = p.add_run()
        run.text = date
        font = run.font
        font.name = 'Yekan Bakh'
        font.bold = True
        font.size = Pt((18*40)/22.2)
        font.color.rgb = RGBColor(73, 31, 103)

    def add_shape1(self, slide):
        shapes = slide.shapes
        left = top = width = height = Inches(1.05)
        shape = shapes.add_shape(
            MSO_SHAPE.RECTANGLE, left, top, width, height
        )
        shape.line.fill.background()

        fill = shape.fill
        fill.solid()
        fill.fore_color.rgb = RGBColor(237, 157, 45)
        shape.left = Inches(0)
        shape.top = Inches(.67)
    def add_shape2(self, slide):
        shapes = slide.shapes
        left = top = Inches(1.0)
        width = Inches(4.75)
        height = Inches(0.6)
        shape = shapes.add_shape(
            MSO_SHAPE.RECTANGLE, left, top, width, height
        )

        fill = shape.fill
        fill.solid()
        fill.fore_color.rgb = RGBColor(73, 31, 103)
        shape.left = Inches(19.25)
        shape.top = Inches(1.1)
        shape.line.fill.background()
        
```

```python
from os import listdir
from os.path import isfile, join, isdir

def all_files(path):
    files = [f for f in listdir(path) if isfile(join(path, f))]
    dirs = [f for f in listdir(path) if isdir(join(path, f))]
    if "Thumbs.db" in files:
        files.remove("Thumbs.db")
    if len(files) == 0:
        return dirs
    else: return files
```

```python
import re
from datetime import datetime

def get_img_data(name_of_file, up_path):
    data = {}
    data['path'] = up_path+ '/' + name_of_file
    indices = [i.start() for i in re.finditer('-', name_of_file)]
    indice_dot = 0
    i = 1
    while indice_dot == 0:        
        if name_of_file[-i] == '.':
            indice_dot = i
        i += 1
    # print(name_of_file, indice_dot)
    id, type, date_string = '','',''
    name_of_file = name_of_file[0: len(name_of_file) - indice_dot]

    id = name_of_file[:name_of_file.index('-')-1]
    content_no_insight = name_of_file[indices[-1] + 1 :]
    # for i in range(indices[0] -1):
    #     id += name_of_file[i]
    
    if not 'insight' in name_of_file:
        
        for i in range(indices[0] + 1, indices[1] - 1):
            type += name_of_file[i]
        for i in range(indices[1] + 1, len(name_of_file)):
            date_string += name_of_file[i]
        if '_' in name_of_file:
            ind_underline = re.finditer('_', name_of_file)

        date_obj = datetime.strptime(date_string.strip(), "%d-%m-%Y").date()

        no = re.findall(r'\d+', type)
        data.update({
            'id' : id,
            'medium': ''.join([i for i in type if not i.isdigit()]).replace('_', '').strip(),
            'type':'Evidence',
            'date': date_obj,
            'content no': int(no[0]),

        })
        if len(no) > 1:
            data['branch'] = str(no[1])
        else:
            data['branch'] = '1'
        if  'live' in data['medium']:
            data['platform'] = 'Twitch'
        elif 'story' in data['medium']:
            data['platform'] = 'Instagram'
        elif 'telegram' in data['medium']:
            data['platform'] = 'Telegram'        
        else:
            data['platform'] = 'YouTube'
            

        return data
    else:
        for i in range(indices[1] + 1, len(name_of_file)):
            type += name_of_file[i]
        content_no = re.findall(r'\d+', content_no_insight)[0]
        data.update({
            'id' : id,
            'medium': ''.join([i for i in type if not i.isdigit()]).replace('_', '').strip(),
            'type':'Insight',
            'date': "",
            'content no': int(content_no),
            'branch':1
        })
        if  'live' in data['medium']:
            data['platform'] = 'Twitch'
        elif 'story' in data['medium']:
            data['platform'] = 'Instagram'
        elif 'telegram' in data['medium']:
            data['platform'] = 'Telegram'
        else:
            data['platform'] = 'YouTube'
        return data
        
get_img_data('xx.x - story10 - 2-1-2023.jpg', '111')
# name[:name.index('-')-1]
```
